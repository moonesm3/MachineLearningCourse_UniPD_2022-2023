# **Perceptron:**

The Perceptron is a Supervised Learning Algorithm for binary classifiers. The Perceptron Model implements the following Algorithm:

1_PbJBdf-WxR0Dd0xHvEoh4A.png

For a particular choice of the weight vector $\boldsymbol{w}$ and bias parameter $\boldsymbol{b}$, the model predicts output $\boldsymbol{\hat{y}}$ for the corresponding input vector $\boldsymbol{x}$.

NOT logical function truth table is of only 1-bit binary input (0 or 1), i.e, the input vector $\boldsymbol{x}$ = A and the corresponding output $\boldsymbol{y}$ = not A

not-table.png

Now for the corresponding weight vector $\boldsymbol{w}$ of the input vector $\boldsymbol{x}$, the associated Perceptron Function can be defined as:

perceptron-math.png

For the implementation, considered weight parameter is $\boldsymbol{w} = -1$ and the bias parameter is $\boldsymbol{b} = 0.5$

# Importing library:

```python
import numpy as np
```

Define functions we need:

```python
#Step function:
def Step(v):
    if v >= 0:
        return 1
    else:
        return 0

#Now we need to design Perceptron model:
#Based on information we know:
def Perceptron(x, w, b):
    v = np.dot(w, x) + b
    y = Step(v)
    return y

#Now design a Not logic function based on w = -1 and b = 0.5
def NotFunction(x):
    w = -1
    b = 0.5
    return Perceptron(x, w, b)
```

```python
#Now is time for testing our model:
TestCase1 = np.array(1)
TestCase2 = np.array(0)

print("NotFunction({}) = {}".format(1, NotFunction(TestCase1)))
print("NotFunction({}) = {}".format(0, NotFunction(TestCase2)))
```

the model predicted output ($\boldsymbol{\hat{y}}$) for each of the test inputs are exactly matched with the NOT logic function conventional output ($\boldsymbol{y}$) according to the truth table.
Hence, it is verified that the perceptron algorithm for NOT logic function is correctly implemented.
